```python
import sympy
#print formatted equations
sympy.init_printing(use_unicode=True)
#setup variables 
R, rho, l, A = sympy.symbols('R rho l A')
#display equation for displacement
R = rho * l / A
graph1 = sympy.plot(R.subs(A,1e-6).subs(rho,1e-6), (l, 0, 1), show=False, line_color='red', label="Area = 1", legend=True, ylabel='R')
graph2 = sympy.plot(R.subs(A,2e-6).subs(rho,1e-6), (l, 0, 1), show=False, line_color='blue', label="Area = 2", legend=True)
graph1.append(graph2[0])
graph1.show()
```
