# Simple Harmonic Motion

## Mathematical description

The equation of motion of a particle undergoing SHM is:  
$ \displaystyle x = A \sin{\left(\omega t \right)} $  

This program will do all of the calculus for us:

```python
#import libraries
from sympy import symbols, pi, sin, cos, init_printing, diff, Eq
from sympy.plotting import plot
#print formatted equations
init_printing(use_unicode=True)
#setup variables 
a, v, x, t, A, f, omega = symbols('a v x t A f omega')
#display equation for displacement
displacement = A * sin(omega * t) 
Eq(x, displacement)
```

In order to get the velocity we need to determine the gradient of the displacement

```python
#differentiate displacement to get velocity
velocity = diff(displacement, t) 
Eq(v, diff(displacement, t))
```

In order to get the acceleration we need to determine the gradient of the velocity

```python
#differentiate velocity to get acceleration
acceleration = diff(velocity, t)
Eq(a, diff(velocity, t))
```

```python
displacementPlot = plot(displacement.subs(A,2).subs(omega,1), (t, 0, 4*pi), show=True, line_color='red', label="Displacement", legend=True)
velocityPlot = plot(velocity.subs(A,2).subs(omega,1), (t, 0, 4*pi), show=True, line_color='green', label="Velocity", legend=True)
accelerationPlot = plot(acceleration.subs(A,2).subs(omega,1), (t, 0, 4*pi), show=True, line_color='blue', label="Acceleration", legend=True)
#displacementPlot.extend(velocityPlot)
#displacementPlot.extend(accelerationPlot)
#displacementPlot.show()
```
